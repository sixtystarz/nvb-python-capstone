```python
import datetime

study_sessions = []

def add_session(sessions):
    """
    Adds a new study session to the list.

    Prompts the user for the subject, hours, and date of the session,
    validates the input, and appends it to the provided sessions list.
    """
    subject = input("Enter subject (e.g., Math, History): ").strip()
    if not subject:
        print("Subject cannot be empty.")
        return

    while True:
        try:
            hours_str = input("Enter hours studied (e.g., 2.5): ")
            hours = float(hours_str)
            if hours <= 0:
                print("Hours must be a positive number.")
                continue
            break
        except ValueError:
            print("Invalid input for hours. Please enter a number.")

    while True:
        date_str = input("Enter date of session (YYYY-MM-DD): ")
        try:

            datetime.datetime.strptime(date_str, "%Y-%m-%d")
            break
        except ValueError:
            print("Invalid date format. Please use YYYY-MM-DD.")

    session = {"subject": subject.capitalize(), "hours": hours, "date": date_str}
    sessions.append(session)
    print(f"Session added: {subject} for {hours} hours on {date_str}")

def view_all_sessions(sessions):
    """
    Displays all recorded study sessions.

    If the sessions list is empty, it informs the user. Otherwise, it prints
    each session's details in a formatted way.
    """
    if not sessions:
        print("No study sessions recorded yet.")
        return

    print("\n--- All Study Sessions ---")
    for i, session in enumerate(sessions):
        print(f"{i+1}. Subject: {session['subject']}, Hours: {session['hours']:.1f}, Date: {session['date']}")
    print("------------------------")

def get_weekly_summary(sessions):
    """
    Calculates and displays the total study hours per subject for the last 7 days.

    It filters sessions from the last week and aggregates hours by subject,
    then prints a summary report.
    """
    if not sessions:
        print("No study sessions recorded yet to generate a weekly summary.")
        return

    today = datetime.date.today()
    seven_days_ago = today - datetime.timedelta(days=7)
    weekly_totals = {}
    found_recent_sessions = False

    for session in sessions:
        session_date = datetime.datetime.strptime(session['date'], "%Y-%m-%d").date()
        if seven_days_ago <= session_date <= today:
            found_recent_sessions = True
            subject = session['subject']
            hours = session['hours']
            weekly_totals[subject] = weekly_totals.get(subject, 0) + hours

    if not found_recent_sessions:
        print("No sessions recorded in the last 7 days.")
        return

    print(f"\n--- Weekly Study Summary (Last 7 Days ending {today}) ---")
    if weekly_totals:
        for subject, total_hours in weekly_totals.items():
            print(f"  {subject}: {total_hours:.1f} hours")
    else:
        print("No sessions found in the last 7 days.")
    print("-------------------------------------------------")

def get_most_studied_subject(sessions):
    """
    Identifies and displays the subject with the most cumulative study hours.

    It aggregates total hours for each subject and determines which subject
    has the highest total, then prints this information.
    """
    if not sessions:
        print("No study sessions recorded yet to determine the most studied subject.")
        return

    subject_hours = {}
    for session in sessions:
        subject = session['subject']
        hours = session['hours']
        subject_hours[subject] = subject_hours.get(subject, 0) + hours

    if not subject_hours:
        print("No subjects found with study hours.")
        return

    most_studied = None
    max_hours = 0
    for subject, total_hours in subject_hours.items():
        if total_hours > max_hours:
            max_hours = total_hours
            most_studied = subject

    print("\n--- Most Studied Subject ---")
    if most_studied:
        print(f"The most studied subject is: {most_studied} with {max_hours:.1f} hours.")
    else:
        print("Could not determine the most studied subject.")
    print("----------------------------")


def main_menu():
    """
    Runs the main interactive menu for the Personal Study Tracker.

    Presents options to the user in a loop, handling their choices to
    add sessions, view sessions, get summaries, or exit the program.
    """
    print("Welcome to the Personal Study Tracker!")
    while True:
        print("\n--- Menu ---")
        print("1. Add Study Session")
        print("2. View All Sessions")
        print("3. Get Weekly Study Summary")
        print("4. Get Most Studied Subject")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            add_session(study_sessions)
        elif choice == '2':
            view_all_sessions(study_sessions)
        elif choice == '3':
            get_weekly_summary(study_sessions)
        elif choice == '4':
            get_most_studied_subject(study_sessions)
        elif choice == '5':
            print("Exiting Study Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")


main_menu()
```
